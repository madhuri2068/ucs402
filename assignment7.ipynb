```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

roll_number = int(input("Enter your roll number: "))
np.random.seed(roll_number)
sales_data = np.random.randint(1000, 5001, size=(12, 4))
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']
df_sales = pd.DataFrame(sales_data, columns=categories, index=months)

print(df_sales.head())
print(df_sales.describe())

df_sales['Total Sales'] = df_sales.sum(axis=1)
df_sales['Growth Rate'] = df_sales['Total Sales'].pct_change() * 100
df_sales['Total Sales per Category'] = df_sales.sum()

df_sales['Avg Growth Electronics'] = df_sales['Electronics'].pct_change() * 100

discount_column = 'Electronics' if roll_number % 2 == 0 else 'Clothing'
df_sales[discount_column] *= 0.9 if discount_column == 'Electronics' else 0.85

plt.figure(figsize=(10, 6))
for col in categories:
    plt.plot(df_sales.index, df_sales[col], marker='o', label=col)
plt.legend()
plt.title("Monthly Sales Trends")
plt.grid()
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=df_sales[categories])
plt.title("Sales Distribution per Category")
```

```python
import numpy as np

array = np.array([[1, -2, 3], [-4, 5, -6]])

abs_array = np.abs(array)
flattened_percentiles = np.percentile(array, [25, 50, 75])
column_percentiles = np.percentile(array, [25, 50, 75], axis=0)
row_percentiles = np.percentile(array, [25, 50, 75], axis=1)

mean_flat, median_flat, std_flat = np.mean(array), np.median(array), np.std(array)
mean_col, median_col, std_col = np.mean(array, axis=0), np.median(array, axis=0), np.std(array, axis=0)
mean_row, median_row, std_row = np.mean(array, axis=1), np.median(array, axis=1), np.std(array, axis=1)

print("Element-wise absolute value:\n", abs_array)
print("\nPercentiles for flattened array:", flattened_percentiles)
print("\nPercentiles for each column:\n", column_percentiles)
print("\nPercentiles for each row:\n", row_percentiles)
print("\nFlattened array - Mean:", mean_flat, "Median:", median_flat, "Std Dev:", std_flat)
print("\nColumn-wise - Mean:", mean_col, "Median:", median_col, "Std Dev:", std_col)
print("\nRow-wise - Mean:", mean_row, "Median:", median_row, "Std Dev:", std_row)
```

```python
import numpy as np

a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

floor_vals = np.floor(a)
ceil_vals = np.ceil(a)
trunc_vals = np.trunc(a)
round_vals = np.round(a, decimals=0)  # Ensuring rounding to the nearest integer

print("Floor values:", floor_vals)
print("Ceil values:", ceil_vals)
print("Truncated values:", trunc_vals)
print("Rounded values:", round_vals)
```
